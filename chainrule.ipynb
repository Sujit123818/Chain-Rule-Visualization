```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python

# Example: z = y^2, y = 2x + 1
# So z = (2x + 1)^2
# Define the functions
```

```python

def y_of_x(x):
    return 2*x + 1

def z_of_y(y):
    return y**2

def z_of_x(x):
    return (2*x + 1)**2

# Derivatives
def dz_dy(y):
    return 2*y

def dy_dx(x):
    return 2

def dz_dx(x):
    return dz_dy(y_of_x(x)) * dy_dx(x)




plt.tight_layout()
plt.savefig('/tmp/chain_rule.png', dpi=150, bbox_inches='tight')
print("\nVisualization saved!")
plt.show()
```

```python
# Test at x = 1
x_test = 1
y_test = y_of_x(x_test)

print("At x = 1:")
print(f"y = {y_test}")
print(f"\nDerivatives:")
print(f"dz/dy = {dz_dy(y_test)}")
print(f"dy/dx = {dy_dx(x_test)}")
print(f"\nChain Rule:")
print(f"dz/dx = (dz/dy) × (dy/dx) = {dz_dy(y_test)} × {dy_dx(x_test)} = {dz_dx(x_test)}")
print(f"\nDirect calculation:")
print(f"dz/dx = d/dx[(2x+1)²] = 2(2x+1)·2 = {2*(2*x_test+1)*2}")
```

```python
# Visualization
x = np.linspace(-2, 3, 200)
y = y_of_x(x)
z = z_of_x(x)

fig, axes = plt.subplots(1, 3, figsize=(14, 4))

# Plot 1: z vs y
axes[0].plot(y, z, 'b-', linewidth=2, label='z = y²')
axes[0].plot(y_test, z_of_y(y_test), 'ro', markersize=10, label=f'Point at y={y_test}')
# Slope at this point
slope_zy = dz_dy(y_test)
axes[0].plot([y_test-1, y_test+1], [z_of_y(y_test) - slope_zy, z_of_y(y_test) + slope_zy], 'r--', linewidth=2, label=f'Slope = dz/dy = {slope_zy}')
axes[0].set_xlabel('y', fontsize=12)
axes[0].set_ylabel('z', fontsize=12)
axes[0].set_title('z = y²\n(Rate: dz/dy)', fontsize=12)
axes[0].grid(True, alpha=0.3)
axes[0].legend()

# Plot 2: y vs x
axes[1].plot(x, y, 'g-', linewidth=2, label='y = 2x + 1')
axes[1].plot(x_test, y_test, 'ro', markersize=10, label=f'Point at x={x_test}')
slope_yx = dy_dx(x_test)
axes[1].plot([x_test-0.5, x_test+0.5], [y_test - slope_yx*0.5, y_test + slope_yx*0.5], 'g--', linewidth=2, label=f'Slope = dy/dx = {slope_yx}')
axes[1].set_xlabel('x', fontsize=12)
axes[1].set_ylabel('y', fontsize=12)
axes[1].set_title('y = 2x + 1\n(Rate: dy/dx)', fontsize=12)
axes[1].grid(True, alpha=0.3)
axes[1].legend()

# Plot 3: z vs x
axes[2].plot(x, z, 'purple', linewidth=2, label='z = (2x+1)²')
axes[2].plot(x_test, z_of_x(x_test), 'ro', markersize=10, label=f'Point at x={x_test}')
slope_zx = dz_dx(x_test)
axes[2].plot([x_test-0.5, x_test+0.5], [z_of_x(x_test) - slope_zx*0.5, z_of_x(x_test) + slope_zx*0.5], 'purple', linestyle='--', linewidth=2, label=f'Slope = dz/dx = {slope_zx}')
axes[2].set_xlabel('x', fontsize=12)
axes[2].set_ylabel('z', fontsize=12)
axes[2].set_title('z = (2x+1)²\n(Combined: dz/dx = dz/dy · dy/dx)', fontsize=12)
axes[2].grid(True, alpha=0.3)
axes[2].legend()
```

```python
#Looking  at the three plots:


Left: \( z \) vs \( y \) — the slope is \( \frac{dz}{dy} = 6 \) (steep!)

Middle: \( y \) vs \( x \) — the slope is \( \frac{dy}{dx} = 2 \) (moderate)

Right: \( z \) vs \( x \) — the slope is \( \frac{dz}{dx} = 12 \) (very steep!)


The chain rule explains that The slope of the final graph (12) is the product of the intermediate slopes: \( 6 \times 2 = 12 \).
```
